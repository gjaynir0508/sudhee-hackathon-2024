```python
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
dic = {
"accident":"Are you a victim or pass by?",
"Victim":"Are you seriously injured?",
"pass by": "Can you upload the pic of the accident?",
"Yes":"Allow us to locate you",
"No":"Can you upload the pic of the accident?"
}
```

```python
print("what is your emergency?")
while 1:
    embedding_1 = model.encode(input(), convert_to_tensor=True)
    sentences = list(dic.keys())
    a = []
    for i in sentences:
        a.append(util.pytorch_cos_sim(embedding_1,model.encode(i,convert_to_tensor=True)))
    print(dic[sentences[np.array(a).argmax()]])
```
